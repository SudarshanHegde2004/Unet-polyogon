```python
# !pip install torch torchvision numpy pillow wandb

# Clean working directory
import shutil
import os

working_directory = '/kaggle/working/'

# Remove all files and folders in the working directory
for item_name in os.listdir(working_directory):
    item_path = os.path.join(working_directory, item_name)
    try:
        if os.path.isfile(item_path) or os.path.islink(item_path):
            os.unlink(item_path)
        elif os.path.isdir(item_path):
            shutil.rmtree(item_path)
    except Exception as error:
        print(f'Failed to delete {item_path}. Reason: {error}')

print("All contents in /kaggle/working/ have been removed.")

# Initialize wandb
import wandb
wandb.login(key='5424a3d65aac1662f5be82d4439aaac35046689e')
print("‚úÖ Wandb authentication successful!")

import os, json, random
from glob import glob
from PIL import Image
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt

# Model configuration
DATA_ROOT = '/kaggle/input/dataset/dataset'
IMAGE_DIM = 128
MINIBATCH_SIZE = 16
TRAINING_EPOCHS = 50
LEARNING_RATE = 1e-4
COMPUTE_DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('üöÄ Using device:', COMPUTE_DEVICE)

# Color definitions
COLOR_DEFINITIONS = {
    'red': [255, 0, 0],
    'green': [0, 255, 0], 
    'blue': [0, 0, 255],
    'yellow': [255, 255, 0],
    'purple': [128, 0, 128],
    'orange': [255, 165, 0],
    'pink': [255, 192, 203],
    'cyan': [0, 255, 255],
    'brown': [165, 42, 42],
    'black': [0, 0, 0]
}

class ShapeColoringDataset(Dataset):
    def __init__(self, root_path, data_split='training'):
        base_path = os.path.join(root_path, data_split)
        data_entries = json.load(open(os.path.join(base_path, 'data.json')))
        self.dataset_samples = []
        
        for entry in data_entries:
            input_path = os.path.join(base_path, 'inputs', entry['input_polygon'])
            output_path = os.path.join(base_path, 'outputs', entry['output_image'])
            color_value = entry['colour'].lower()
            self.dataset_samples.append((input_path, output_path, color_value))
            
        unique_colors = list(set([sample[2] for sample in self.dataset_samples]))
        self.color_mapping = {color: idx for idx, color in enumerate(unique_colors)}
        self.color_count = len(unique_colors)
        
        self.image_transform = transforms.Compose([
            transforms.Resize((IMAGE_DIM, IMAGE_DIM)),
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.dataset_samples)

    def __getitem__(self, index):
        input_path, output_path, color_value = self.dataset_samples[index]
        
        input_image = Image.open(input_path).convert('RGB')
        input_tensor = self.image_transform(input_image)
        
        target_image = Image.open(output_path).convert('RGB')
        target_tensor = self.image_transform(target_image)
        
        target_gray = transforms.Grayscale()(target_tensor)
        fill_mask = (target_gray < 0.95).float()
        
        color_index = self.color_mapping[color_value]
        
        if color_value in COLOR_DEFINITIONS:
            rgb_values = torch.tensor(COLOR_DEFINITIONS[color_value], dtype=torch.float32) / 255.0
        else:
            target_array = np.array(target_image)
            colored_pixels = target_array[target_array.mean(axis=2) < 240]
            if len(colored_pixels) > 0:
                rgb_values = torch.tensor(colored_pixels.mean(axis=0), dtype=torch.float32) / 255.0
            else:
                rgb_values = torch.tensor([1.0, 0.0, 0.0])
        
        return input_tensor, fill_mask, color_index, rgb_values

class ColorAwareUNet(nn.Module):
    def __init__(self, num_colors, channels=[64, 128, 256, 512]):
        super().__init__()
        
        self.color_embed = nn.Embedding(num_colors, 64)
        self.color_project = nn.Linear(64, IMAGE_DIM * IMAGE_DIM)
        
        self.encoder_blocks = nn.ModuleList()
        input_channels = 4
        for channel in channels:
            self.encoder_blocks.append(nn.Sequential(
                nn.Conv2d(input_channels, channel, 3, padding=1), 
                nn.BatchNorm2d(channel),
                nn.ReLU(inplace=True),
                nn.Conv2d(channel, channel, 3, padding=1),     
                nn.BatchNorm2d(channel),
                nn.ReLU(inplace=True)
            ))
            input_channels = channel
            
        self.downsample = nn.MaxPool2d(2)
        
        self.middle_layer = nn.Sequential(
            nn.Conv2d(channels[-1], channels[-1]*2, 3, padding=1), 
            nn.BatchNorm2d(channels[-1]*2),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Conv2d(channels[-1]*2, channels[-1]*2, 3, padding=1), 
            nn.BatchNorm2d(channels[-1]*2),
            nn.ReLU(inplace=True)
        )
        
        reversed_channels = channels[::-1]
        self.upscale_layers = nn.ModuleList([
            nn.ConvTranspose2d(ch*2, ch, 2, stride=2) for ch in reversed_channels
        ])
        self.decoder_blocks = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(ch*2, ch, 3, padding=1), 
                nn.BatchNorm2d(ch),
                nn.ReLU(inplace=True),
                nn.Conv2d(ch, ch, 3, padding=1),   
                nn.BatchNorm2d(ch),
                nn.ReLU(inplace=True)
            ) for ch in reversed_channels
        ])
        
        self.final_layer = nn.Conv2d(channels[0], 1, 1)

    def forward(self, image_input, color_input):
        batch_size = image_input.size(0)
        
        color_encoded = self.color_embed(color_input)
        color_features = self.color_project(color_encoded)
        color_features = color_features.view(batch_size, 1, IMAGE_DIM, IMAGE_DIM)
        
        combined_input = torch.cat([image_input, color_features], dim=1)
        
        skip_connections = []
        for encoder in self.encoder_blocks:
            combined_input = encoder(combined_input)
            skip_connections.append(combined_input)
            combined_input = self.downsample(combined_input)
            
        combined_input = self.middle_layer(combined_input)
        
        for upscale, decoder, skip in zip(self.upscale_layers, self.decoder_blocks, reversed(skip_connections)):
            combined_input = upscale(combined_input)
            if combined_input.shape[2:] != skip.shape[2:]:
                combined_input = F.interpolate(combined_input, size=skip.shape[2:], mode='bilinear', align_corners=False)
            combined_input = torch.cat([skip, combined_input], dim=1)
            combined_input = decoder(combined_input)
            
        return self.final_layer(combined_input)

def log_model_predictions(model, dataset, color_encoder, current_epoch, sample_count=6):
    model.eval()
    sample_indices = random.sample(range(len(dataset)), min(sample_count, len(dataset)))
    prediction_images = []
    
    with torch.no_grad():
        for idx, sample_idx in enumerate(sample_indices):
            input_img, target_mask, color_idx, color_rgb = dataset[sample_idx]
            input_batch = input_img.unsqueeze(0).to(COMPUTE_DEVICE)
            color_batch = torch.tensor([color_idx]).to(COMPUTE_DEVICE)
            
            output_logits = model(input_batch, color_batch)[0, 0].cpu()
            predicted_mask = (torch.sigmoid(output_logits) > 0.5).numpy()
            
            fig, ax = plt.subplots(1, 4, figsize=(12, 3))
            
            input_array = input_img.permute(1, 2, 0).numpy()
            ax[0].imshow(input_array)
            ax[0].set_title('Input')
            ax[0].axis('off')
            
            ax[1].imshow(target_mask[0].numpy(), cmap='gray')
            ax[1].set_title('Target Mask')
            ax[1].axis('off')
            
            ax[2].imshow(predicted_mask, cmap='gray')
            ax[2].set_title('Predicted Mask')
            ax[2].axis('off')
            
            color_name = list(color_encoder.keys())[list(color_encoder.values()).index(color_idx)]
            fill_color = COLOR_DEFINITIONS.get(color_name, [255, 0, 0])
            
            input_uint8 = (input_array * 255).astype(np.uint8)
            colored_output = input_uint8.copy()
            colored_output[predicted_mask] = fill_color
            
            ax[3].imshow(colored_output)
            ax[3].set_title(f'Colored ({color_name})')
            ax[3].axis('off')
            
            plt.tight_layout()
            
            wandb_img = wandb.Image(fig, caption=f"Sample {idx+1}: {color_name} - Epoch {current_epoch}")
            prediction_images.append(wandb_img)
            plt.close(fig)
    
    wandb.log({f"predictions/epoch_{current_epoch}": prediction_images}, step=current_epoch)
    print(f"üì∏ Logged {len(prediction_images)} prediction samples to wandb")

def execute_training():
    training_run = wandb.init(
        project="polygon-coloring-network",
        name=f"unet-lr{LEARNING_RATE}-bs{MINIBATCH_SIZE}-ep{TRAINING_EPOCHS}",
        config={
            "learning_rate": LEARNING_RATE,
            "batch_size": MINIBATCH_SIZE,
            "epochs": TRAINING_EPOCHS,
            "image_size": IMAGE_DIM,
            "model": "ColorAwareUNet",
            "optimizer": "Adam",
            "loss": "BCEWithLogitsLoss",
            "device": str(COMPUTE_DEVICE)
        }
    )
    
    train_data = ShapeColoringDataset(DATA_ROOT, 'training')
    val_data = ShapeColoringDataset(DATA_ROOT, 'validation')
    
    train_loader = DataLoader(train_data, MINIBATCH_SIZE, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_data, MINIBATCH_SIZE, shuffle=False, num_workers=2)
    
    wandb.log({
        "data/train_count": len(train_data),
        "data/val_count": len(val_data),
        "data/color_count": train_data.color_count,
        "data/colors": list(train_data.color_mapping.keys())
    })
    
    model = ColorAwareUNet(train_data.color_count).to(COMPUTE_DEVICE)
    
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    
    wandb.log({
        "model/params_total": total_params,
        "model/params_trainable": trainable_params,
        "model/size_mb": total_params * 4 / (1024 * 1024)
    })
    
    wandb.watch(model, log="all", log_freq=100)
    
    print(f"‚úÖ Model initialized with {total_params:,} parameters")
    print(f"üìä Training samples: {len(train_data)} | Validation samples: {len(val_data)}")
    
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)
    lr_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5)
    loss_function = nn.BCEWithLogitsLoss()
    best_loss = float('inf')
    
    workspace_model_path = os.path.join(working_directory, 'best_model_weights.pth')
    
    for epoch in range(1, TRAINING_EPOCHS + 1):
        model.train()
        epoch_loss = 0
        epoch_accuracy = 0
        
        for batch_num, (inputs, masks, colors, _) in enumerate(train_loader):
            inputs, masks, colors = inputs.to(COMPUTE_DEVICE), masks.to(COMPUTE_DEVICE), colors.to(COMPUTE_DEVICE)
            
            outputs = model(inputs, colors)
            loss = loss_function(outputs, masks)
            
            with torch.no_grad():
                predictions = torch.sigmoid(outputs) > 0.5
                accuracy = (predictions == masks.bool()).float().mean()
                epoch_accuracy += accuracy.item()
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()
            
            if batch_num % 50 == 0:
                wandb.log({
                    "batch/loss": loss.item(),
                    "batch/accuracy": accuracy.item(),
                    "batch/lr": optimizer.param_groups[0]['lr'],
                    "batch/epoch": epoch,
                    "batch/number": batch_num
                })
        
        epoch_loss /= len(train_loader)
        epoch_accuracy /= len(train_loader)
        
        model.eval()
        val_loss = 0
        val_accuracy = 0
        
        with torch.no_grad():
            for inputs, masks, colors, _ in val_loader:
                inputs, masks, colors = inputs.to(COMPUTE_DEVICE), masks.to(COMPUTE_DEVICE), colors.to(COMPUTE_DEVICE)
                
                outputs = model(inputs, colors)
                loss = loss_function(outputs, masks)
                
                predictions = torch.sigmoid(outputs) > 0.5
                accuracy = (predictions == masks.bool()).float().mean()
                
                val_loss += loss.item()
                val_accuracy += accuracy.item()
        
        val_loss /= len(val_loader)
        val_accuracy /= len(val_loader)
        
        lr_scheduler.step(val_loss)
        
        wandb.log({
            "epoch/train_loss": epoch_loss,
            "epoch/val_loss": val_loss,
            "epoch/train_acc": epoch_accuracy,
            "epoch/val_acc": val_accuracy,
            "epoch/lr": optimizer.param_groups[0]['lr'],
            "epoch": epoch
        })
        
        print(f"Epoch {epoch:2d}/{TRAINING_EPOCHS} | Loss: {epoch_loss:.4f} | Val Loss: {val_loss:.4f} | "
              f"Acc: {epoch_accuracy:.4f} | Val Acc: {val_accuracy:.4f}")
        
        if val_loss < best_loss:
            best_loss = val_loss
            
            torch.save({
                'state_dict': model.state_dict(),
                'color_encoder': train_data.color_mapping,
                'color_count': train_data.color_count,
                'epoch': epoch,
                'val_loss': val_loss
            }, workspace_model_path)
            
            model_artifact = wandb.Artifact(
                name="top_performing_model", 
                type="model",
                description=f"Best model at epoch {epoch} with val_loss {val_loss:.4f}"
            )
            model_artifact.add_file(workspace_model_path)
            wandb.log_artifact(model_artifact)
            
            wandb.log({"best_val_loss": best_loss, "best_epoch": epoch})
        
        if epoch % 5 == 0:
            log_model_predictions(model, val_data, train_data.color_mapping, epoch)
    
    wandb.run.summary["final_train_loss"] = epoch_loss
    wandb.run.summary["final_val_loss"] = val_loss
    wandb.run.summary["best_val_loss"] = best_loss
    wandb.run.summary["total_epochs"] = TRAINING_EPOCHS
    
    print("‚úÖ Training completed!")
    print(f"üèÜ Best validation loss: {best_loss:.4f}")
    print(f"üíæ Model saved to: {workspace_model_path}")
    print(f"üîó View results: https://wandb.ai/{wandb.api.default_entity}/polygon-coloring-network")
    
    wandb.finish()
    return model, train_data.color_mapping, workspace_model_path

print("üéØ Starting training with wandb monitoring...")
trained_model, color_encoding, model_path = execute_training()

def compute_metrics(prediction, ground_truth, threshold=0.5):
    binary_pred = (prediction > threshold).astype(np.uint8)
    binary_gt = ground_truth.astype(np.uint8)
    
    accuracy = np.mean(binary_pred == binary_gt)
    
    intersect = np.sum(binary_pred * binary_gt)
    union_area = np.sum(binary_pred) + np.sum(binary_gt) - intersect
    iou = intersect / (union_area + 1e-8)
    
    return accuracy, iou

def evaluate_model(model_file=None, data_split='validation', num_examples=6):
    if model_file is None:
        model_file = os.path.join(working_directory, 'best_model_weights.pth')
    
    eval_run = wandb.init(project="polygon-coloring-network", job_type="evaluation")
    
    print("üìù Loading model for evaluation...")
    
    checkpoint = torch.load(model_file, map_location=COMPUTE_DEVICE)
    model = ColorAwareUNet(checkpoint['color_count']).to(COMPUTE_DEVICE)
    model.load_state_dict(checkpoint['state_dict'])
    model.eval()
    
    color_encoder = checkpoint['color_encoder']
    
    data_path = os.path.join(DATA_ROOT, data_split)
    data_entries = json.load(open(os.path.join(data_path, 'data.json')))
    
    test_samples = random.sample(range(len(data_entries)), min(num_examples, len(data_entries)))
    
    total_accuracy = 0
    total_iou = 0
    result_images = []
    
    print(f"üß™ Evaluating on {num_examples} samples...")
    
    with torch.no_grad():
        for idx, sample_idx in enumerate(test_samples):
            sample = data_entries[sample_idx]
            input_file = os.path.join(data_path, 'inputs', sample['input_polygon'])
            output_file = os.path.join(data_path, 'outputs', sample['output_image'])
            color_name = sample['colour'].lower()
            
            input_img = Image.open(input_file).convert('RGB').resize((IMAGE_DIM, IMAGE_DIM))
            input_tensor = transforms.ToTensor()(input_img).unsqueeze(0).to(COMPUTE_DEVICE)
            
            color_idx = torch.tensor([color_encoder[color_name]]).to(COMPUTE_DEVICE)
            
            logits = model(input_tensor, color_idx)[0, 0].cpu()
            pred_mask = torch.sigmoid(logits).numpy()
            
            gt_img = Image.open(output_file).convert('RGB').resize((IMAGE_DIM, IMAGE_DIM))
            gt_gray = np.array(Image.open(output_file).convert('L').resize((IMAGE_DIM, IMAGE_DIM)))
            gt_mask = (gt_gray < 240).astype(np.float32)
            
            input_array = np.array(input_img)
            colored_output = input_array.copy()
            
            fill_color = COLOR_DEFINITIONS.get(color_name, [255, 0, 0])
            colored_output[pred_mask > 0.5] = fill_color
            
            accuracy, iou = compute_metrics(pred_mask, gt_mask)
            total_accuracy += accuracy
            total_iou += iou
            
            fig, ax = plt.subplots(1, 4, figsize=(16, 4))
            
            ax[0].imshow(input_img)
            ax[0].set_title(f'Input\n({color_name})')
            ax[0].axis('off')
            
            ax[1].imshow(gt_img)
            ax[1].set_title('Ground Truth')
            ax[1].axis('off')
            
            ax[2].imshow(colored_output)
            ax[2].set_title(f'Prediction\nIoU: {iou:.3f}')
            ax[2].axis('off')
            
            mask_comparison = np.hstack([gt_mask, pred_mask > 0.5])
            ax[3].imshow(mask_comparison, cmap='gray')
            ax[3].set_title(f'GT | Pred Masks\nAcc: {accuracy:.3f}')
            ax[3].axis('off')
            
            plt.tight_layout()
            plt.show()
            
            result_images.append(wandb.Image(fig, caption=f"Test {idx+1}: {color_name}"))
            plt.close()
            
            print(f"Sample {idx+1}: {color_name} - IoU: {iou:.3f}, Accuracy: {accuracy:.3f}")
    
    avg_accuracy = total_accuracy / num_examples
    avg_iou = total_iou / num_examples
    
    wandb.log({
        "eval/avg_accuracy": avg_accuracy,
        "eval/avg_iou": avg_iou,
        "eval/results": result_images
    })
    
    print("\n" + "="*50)
    print("üéØ EVALUATION SUMMARY")
    print("="*50)
    print(f"Samples Evaluated: {num_examples}")
    print(f"Average Accuracy: {avg_accuracy:.4f}")
    print(f"Average IoU: {avg_iou:.4f}")
    print("="*50)
    
    wandb.finish()
    return avg_accuracy, avg_iou

print("üß™ Starting model evaluation...")
final_accuracy, final_iou = evaluate_model(
    model_file=None,
    data_split='validation', 
    num_examples=8
)
```
